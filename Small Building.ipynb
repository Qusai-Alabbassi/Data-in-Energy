```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# First, we have to upload our local excel sheet
```

```python
file_path = r'/Users/qusaialabbassi/Desktop/pythonprojects/Energy Management/building_energy_data.xlsx'
```

```python
df = pd.read_excel(file_path)
```

```python
# Now, we have to convert the timestamp (define) in our excel sheet to a readable version by Python
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
df['Hour']=df['Timestamp'].dt.hour
```

```python
# Now, we have to define our system "Small Buidling"
```

```python
Systems = ['HVAC (Wh)','Lighting (Wh)','Plugs (Wh)']
```

```python
# Now, we have to calcualte the monthly energy in kWh Assuming the cost for kWh is 0.12$
```

```python
monthly_kwh=df[Systems].sum()/1000
```

```python
monthly_cost=monthly_kwh*0.12
```

```python
# Now, the plotting section :) :)
```

```python
monthly_cost.plot(kind='bar', title='Monthly Energy Cost Per System ($)')
plt.ylabel('USD')
plt.grid(True)
plt.tight_layout()
plt.show()
```
